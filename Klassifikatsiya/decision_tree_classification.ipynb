```python
# Kerakli kutubxonalarni chaqirib olamiz
from sklearn import tree  # Decision Tree uchun
from sklearn.model_selection import train_test_split  # Malumotlarni bolish uchun
from sklearn.metrics import accuracy_score  # aniqlikni olchash uchun
```

```python

# X - xususiyatlar, Y - natija
#0 = yashil, 1 = qizil
X = [
    [150, 0],
    [170, 0],
    [140, 1],
    [130, 1],
    [180, 0],
    [160, 1]
]

#  0 = olma, 1 = nok
Y = [0, 0, 0, 0, 1, 1]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=1)
```

```python
# Decision Tree modelini yaratamiz

model = tree.DecisionTreeClassifier()
```

```python
model.fit(X_train, Y_train)
```

```python
Y_pred = model.predict(X_test)
```

```python
accuracy = accuracy_score(Y_test, Y_pred)
print("Modelning aniqligi :", accuracy)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,7))
tree.plot_tree(model, filled=True, feature_names=["Og'irlik", "Rang"], class_names=["Olma", "Nok"])
plt.show()
```
