```python
import numpy as np
from sklearn.svm import SVC
```

```python
# 0 = yashil, 1 = qizil
X = [
    [150, 0],
    [170, 0],
    [140, 1],
    [130, 1],
    [180, 0],
    [160, 1]
]

# 0 = olma, 1 = nok
Y = [0, 0, 0, 0, 1, 1]
```

```python
# Numpy arrayga o‘tkazamiz
X = np.array(X)
Y = np.array(Y)
```

```python
model = SVC(kernel='rbf', C=1.0, probability=True)
model.fit(X, Y)
```

```python
test = np.array([[145, 1]])
prediction = model.predict(test)
proba = model.predict_proba(test)
```

```python
print("Bashorat:", prediction[0])  # 0 yoki 1
print("Ehtimollar [olma, nok]:", proba[0])  # masalan [0.7 0.3]
```

```python
x1 = X[:, 0]
x2 = X[:, 1]
colors = ['green' if y == 0 else 'red' for y in Y]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(x1, x2, c=colors, label='O‘quv ma’lumotlari')
plt.scatter(145, 1, c='blue', marker='x', s=200, label='Test: [145, 1]')
```
